```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.offsetbox import OffsetImage,AnnotationBbox
plt.style.use('ggplot')
```

```python
data = pd.read_excel("rp_cricket_stats.xlsx")
```

```python
data.head()
```

```python
matches = []
for i in data["match"]:
    if i not in matches:
        matches.append(i)
matches
```

```python
yash = []
yash_data = data[data["name"] == "yash"]
for i in yash_data["score"]:
    yash.append(i)
yash
```

```python
mayur = []
mayur_data = data[data["name"] == "mayur"]
for i in mayur_data["score"]:
    mayur.append(i)
mayur
```

```python
shashank = []
shashank_data = data[data["name"] == "shashank"]
for i in shashank_data["score"]:
    shashank.append(i)
shashank
```

```python
pari = []
pari_data = data[data["name"] == "pari"]
for i in pari_data["score"]:
    pari.append(i)
pari
```

```python
anuj = []
anuj_data = data[data["name"] == "anuj"]
for i in anuj_data["score"]:
    anuj.append(i)
anuj
```

### Total Runs Comparison

```python
runs = [sum(yash),sum(mayur),sum(shashank),sum(pari),sum(anuj)]
names = ["yash","mayur","shashank","pari","anuj"]
runs
```

```python
fig = plt.figure(figsize = (10, 5))

def get_img(name):
    path = "web_project/profile_pics/{}.jpeg".format(name.title())
    im = plt.imread(path)
    return im

def offset_image(coord, name, ax):
    img = get_img(name)
    im = OffsetImage(img, zoom=0.025)
    im.image.axes = ax

    ab = AnnotationBbox(im, (coord, 0),  xybox=(0., -16.), frameon=False,
                        xycoords='data',  boxcoords="offset points", pad=0)

    ax.add_artist(ab)
    

fig, ax = plt.subplots()

ax.bar(range(len(names)), runs, width=0.5,align="center",color="yellow")
ax.set_xticks(range(len(names)))
ax.set_xticklabels(names)
ax.tick_params(axis='x', which='major', pad=26)

for i, c in enumerate(names):
    offset_image(i, c, ax)
    
plt.ylabel("Total Runs")
plt.title("Total Runs Comparison")

plt.show()

X = np.array(fig.canvas.renderer.buffer_rgba())
plt.imsave("web_project/results/totalruns.png",X)
```

### Match wise comparison

```python
angles=np.linspace(0,2*np.pi,len(matches), endpoint=False)
angles=np.concatenate((angles,[angles[0]]))
```

```python
matches.append(matches[0])
yash.append(yash[0])
mayur.append(mayur[0])
shashank.append(shashank[0])
pari.append(pari[0])
anuj.append(anuj[0])
```

```python
fig=plt.figure(figsize=(7,7))
ax=fig.add_subplot(111, polar=True)

#yash Plot
ax.plot(angles,yash, 'o-', color='blue', linewidth=2, label=yash)
ax.fill(angles,yash, alpha=0.25, color='blue')
#mayur Plot
ax.plot(angles,mayur, 'o-', color='orange', linewidth=2, label=mayur)
ax.fill(angles,mayur, alpha=0.25, color='orange')
#shashank Plot
ax.plot(angles,shashank, 'o-', color='red', linewidth=2, label=shashank)
ax.fill(angles,shashank, alpha=0.25, color='red')
#pari Plot
ax.plot(angles,pari, 'o-', color='yellow', linewidth=2, label=pari)
ax.fill(angles,pari, alpha=0.25, color='yellow')
#anuj Plot
ax.plot(angles,anuj, 'o-', color='black', linewidth=2, label=anuj)
ax.fill(angles,anuj, alpha=0.25, color='black')

match = ["match1","match2","match3","match4","match5","match6","match7","match8","match9","match10",
         "match11","match12","match13","match14","match15","match16","match17","match18"]

ax.set_rgrids([0,10,20,30,40,50])
ax.set_thetagrids(angles * 180/np.pi, match)
ax.set_title("Rp cricket", weight='bold', size='medium', position=(0.5, 1.1),
                     horizontalalignment='center', verticalalignment='center')

plt.grid(True)
plt.tight_layout()
labels = ("yash","mayur","shashank","pari","anuj")
plt.legend(labels, loc=(0.9, .95),labelspacing=0.1, fontsize='medium')
plt.show()

X = np.array(fig.canvas.renderer.buffer_rgba())
plt.imsave("web_project/results/matchwise.png",X)
```
